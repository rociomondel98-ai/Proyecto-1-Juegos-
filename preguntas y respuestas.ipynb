PREGUNTAS Y RESPUESTAS

```python
# BIENVENIDA
nombre_jugador = input("Indícame tu nombre de jugadora: ")
print(f"Hola {nombre_jugador.upper()}, Bienvenida al juego de trivia")
#PREGUNTAS PARA EL JUEGO 
preguntas = [ {"pregunta": "¿El sol es una estrella? (si/no): ", "respuesta": "SI"}, #lista de diccionarios para preguntas y respuestas
    {"pregunta": "¿La capital de Francia es Berlín? (si/no): ", "respuesta": "NO"},
    {"pregunta": "¿Python es un lenguaje de programación? (si/no): ", "respuesta": "SI"}]
# Función de la ronda de preguntas al jugador
def hacer_pregunta(pregunta_dict):
    respuesta_usuario = input(pregunta_dict["pregunta"]).strip().upper() 
    if respuesta_usuario == pregunta_dict["respuesta"]:
        print("✅ Correcto\n")
        return True
    else:
        print("❌ Incorrecto.\n") # (\n) para dar espacio en la visual
        return False
# Contador de respuestas correctas  
puntos = 0
# Bucle principal: recorre las preguntas
for pregunta in preguntas:
    if hacer_pregunta(pregunta): #funcion
        puntos += 1 
# Mostrar resultado final
print("Juego terminado")
print(f"Respuestas correctas: {puntos} de {len(preguntas)}")



```
