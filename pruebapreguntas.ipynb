```python
#juego de preguntas y respuestas
#el usuario es el jugador 1 que es el que va a responder a las preguntas
#el jugador 2 es la computadora que pregunta, y checkea si la respuesta es correcta
#va a haber tres categorÃ­as: historia, geografÃ­a y literatura.
#y dentro de cada categorÃ­a preguntas fÃ¡ciles o difÃ­ciles 
#el objetivo es conseguir cinco puntos. si se contesta bien a una pregunta difÃ­cil se suman dos mientras que si lo hace a una facil es un punto 
#tiene 3 intentos, si responde 3 mal se acaba el juego 
```

```python
#importamos random para que elija las opciones del diccionario 
import random 
#creamos diccionarios con preguntas 
diccionariopreguntas_geografiaf = {'Â¿En quÃ© continente se encuentra Portugal?' : 'Europa', 'Â¿CuÃ¡l es el rÃ­o mÃ¡s largo del mundo?' : 'Amazonas', 'Â¿CuÃ¡l es la capital de Irlanda?' : 'DublÃ­n'}
diccionariopreguntas_geografiad = {'Â¿CuÃ¡l es el paÃ­s con mÃ¡s husos horarios en su territorio?' : 'Francia', 'Â¿QuÃ© ciudad es la capital mÃ¡s alta del mundo?' : 'La Paz', 'Â¿CÃ³mo se llama el estrecho que separa EspaÃ±a de Marruecos?' : 'Estrecho de Gibraltar'}
diccionariopreguntas_historiaf = {'Â¿QuiÃ©n fue derrotado en Waterloo?' : 'NapoleÃ³n', 'CÃ³mo se llamaba el barco que se hundiÃ³ en 1912?' : 'Titanic', 'Â¿QuÃ© civilizaciÃ³n hablaba latÃ­n?' : 'Romanos'}
diccionariopreguntas_historiad = {'Â¿En quÃ© aÃ±o cayÃ³ el Muro de BerlÃ­n?' : '1989', 'Â¿CÃ³mo se llamaba el primer presidente de los EEUU?' : 'George Washington', 'Â¿CuÃ¡l era el nombre de la ruta comercial que conectaba China con Europa?': 'Ruta de la Seda'}
diccionariopreguntas_literaturaf = {'Â¿QuiÃ©n escribiÃ³ "La Casa de Bernarda Alba"?' : 'Lorca', 'Â¿QuiÃ©n escribiÃ³ "Hamlet"?' : 'Shakespeare', 'Â¿CÃ³mo se llama el protagonista del libro "El SeÃ±or de los Anillos?' : 'Frodo'}
diccionariopreguntas_literaturad = {'Â¿CÃ³mo se llama el monstruo que creÃ³ Mary Shelley en su novela?' : 'Frankenstein', 'Â¿QuiÃ©n escribiÃ³ "1984"?' : 'George Orwell', 'Â¿DÃ³nde naciÃ³ Pablo Neruda?' : 'Chile'}
#vidas y recompensas 
total_recompensa = 5
total_vidas = 0
inicio_recompensa = 0 
inicio_vidas = 3
recompensa1 = 'ðŸ”“' #para cuando la respuesta sea correcta 
recompensa2 = 'ðŸ”’' #para cuando sea incorrecta 
recompensa3 = 'ðŸ‘‘'
#creamos mensaje de bienvenida y explicaciÃ³n de las reglas
jugadora = input('Â¿CÃ³mo te llamas?')
bienvenida = print('Â¡Bienvenida', jugadora,'al juego de Preguntas y Respuestas!')
bienvenida2 = print('PodrÃ¡s elegir la categorÃ­a de las preguntas y la dificultad. Â¡Ojo! Tu objetivo serÃ¡ conseguir cinco puntos: las respuestas correctas a las preguntas fÃ¡ciles te suman un punto y las difÃ­ciles dos.')
bienvenida3 = print('Pero tambiÃ©n tenemos un sistema de vidas. ContarÃ¡s con 3, si estas llegan a cero, se terminarÃ¡ el juego')
bienvenida4 = print('Si estÃ¡s preparada, elije una categorÃ­a y una dificultad. Â¡Suerte!')
#creamos bucle para que el juego no se ejecute solo una vez
while inicio_vidas > total_vidas and inicio_recompensa < total_recompensa:
#creamos la pregunta para que el usuario elija
    opciones = ['GeografÃ­a', 'Historia', 'Literatura']
    opciones2 = ['FÃ¡cil', 'DifÃ­cil']
    usuario = input('Elige una de las siguientes categorÃ­as: GeografÃ­a, Historia o Literatura').capitalize()
    usuario2 = input(f'Has elegido {usuario}.Dentro de esta categorÃ­a, Â¿quieres preguntas fÃ¡ciles o difÃ­ciles?').capitalize()
#acotamos las respuestas que puede dar el usuario
    if usuario in opciones:
        if usuario2 in opciones2:
            print(usuario)
            print(usuario2)
        else: 
            pass
    elif usuario != opciones:
        print('Â¡Tienes que elegir entre GeografÃ­a, Historia o Literatura!')
    else:
        pass 
#creamos condiciones para que el random genere la pregunta que el jugador haya elegido 
    if usuario == 'GeografÃ­a':
        if usuario2 == 'FÃ¡cil':
            dict_actual = diccionariopreguntas_geografiaf
        else: 
            usuario2 == 'DifÃ­cil'
            dict_actual = diccionariopreguntas_geografiad
    elif usuario == 'Historia': 
        if usuario2 == 'FÃ¡cil':
            dict_actual = diccionariopreguntas_historiaf
        else:
            usuario2 == 'DifÃ­cil'
            dict_actual = diccionariopreguntas_historiad
    elif usuario == 'Literatura':
        if usuario2 == 'FÃ¡cil':
            dict_actual = diccionariopreguntas_literaturaf
        else:   
            usuario2 == 'DifÃ­cil'
            dict_actual = diccionariopreguntas_literaturad
    else:
        pass 
#creamos condiciones para cada respuesta y lo que significa en la partida
    pregunta = random.choice(list(dict_actual.keys()))
    respuesta_correcta = dict_actual[pregunta]
    print('La pregunta es:', pregunta)
    respuesta_usuario = input('Escribe tu respuesta:').capitalize()
    if respuesta_usuario == respuesta_correcta.capitalize():
        if usuario2 == 'FÃ¡cil':
            print('Â¡GENIAL! Tu respuesta es correcta, has desbloqueado', recompensa1, 'una', recompensa3)
            inicio_recompensa += 1
        elif usuario2 == 'DifÃ­cil':
            print('Â¡INCREÃBLE! Por ser una pregunta difÃ­cil has desbloqueado', recompensa1, 'dos', recompensa3)
        else: 
            pass
    else:
        inicio_vidas -= 1
        print('Â¡Oh no! La respuesta correcta era:', respuesta_correcta,'. No has desbloqueado ninguna recompensa', recompensa2)
        print('Te quedan', inicio_vidas,'vidas.')

#acotamos contadores de vida y final de partida 
if inicio_vidas == total_vidas:
    print('Se acabÃ³ el juego',jugadora,'. Cuando quieras volver a probar tu conocimiento vuelve por aquÃ­!')
elif inicio_recompensa == total_recompensa: 
    print('Â¡Enhorabuena',jugadora,'has ganado!')
else:
    pass 
```
