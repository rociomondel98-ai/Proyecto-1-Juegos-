Working Agreement

1. comunicacion clara y respetuosa. No hay excusas, entre todas buscamos la solucion.
2. hacemos dos ejercicios bien ejecutados y si sobra tiempo los pulimos.
3. nos dividimos en dos grupos.un grupo hará el juego piedra,papel o tijera y el otro grupo preguntas y respuestas.
4. se harán reuniones regurales tanto entre los subgupos como el equipo entero para contarstar ideas y así detectar puntos de mejora.
5. se hará una demo de los juegos
6. se mantendrá una actidud positiva durante la elaboración del proyecto.
