```python
import random

#-----------RECURSOS PREVIOS A LAS RONDAS-----------
opciones = ("Piedra", "Papel", "Tijera")  # OPCIONES PARA QUE EL RAMDON ELIGA
opcion_correcta = {"Piedra": "Tijera","Papel": "Piedra","Tijera": "Papel"} # REGLAS DEL JUEGO / {CLAVE=GANA:VALOR=PIERDE}

victorias_jugadora = 0 # VARIABLES DE CONTROL
victorias_pc = 0       # VARIABLES DE CONTROL
puntos_totales = 3     # VARIABLES DE CONTROL
contador_rondas = 0    # VARIABLES DE CONTROL
nombre_jugadora = input("Para iniciar, indÃ­came tu nombre de jugadora: ").upper() # VARIABLE: NOMBRE JUGADORA


#----- FUNCIÃ“N: VALIDACIÃ“N Y CONTROL DE ERRORES -----
def eleccion_jugadora(): 
    while True: #WHILE TRUE YA QUE NECESITO QUE EL BUCLE ESTÃ‰ ACTIVO HASTA QUE LA JUGADORA ESCRIBA SU OPCION Y SEA CORRECTA GRAMATICALMENTE.
        jugadora_elige = input("Elige piedra, papel o tijera: ").title().strip() # PREGUNTA 
        if jugadora_elige in opciones: #SI LA ELECCIÃ“N DE LA JUGADORA ESTÃ EN LAS OPCIONES 
            return jugadora_elige #RETORNAME SU OPCIÃ“N
        
        else:  #SINO, INDICALE QUE HA ESCRITO MAL LA PALABRA
            print("âš ï¸ Error. Debes elegir entre Piedra, Papel o Tijera.\n") #MUESTRALE MSJ ERROR


#------------------- BIENVENIDA -------------------- 
print(f"Bienvenida al juego {nombre_jugadora}, vamos a comenzar.\n") # BIENVENIDA AL JUEGO
print("Debes elegir entre: Piedra âœŠğŸ» / Papel âœ‹ğŸ» / Tijera âœŒğŸ»")       # REGLAS DEL JUEGO
print("Para vencer a ğŸ¤– tienes que ganar tres rondas.\n")            #PARAMETROS PARA GANAR


#------------ BUCLE CONTROL DE RONDAS -------------
while victorias_jugadora < puntos_totales and victorias_pc < puntos_totales: #MIENTRAS LAS VISTORIAS DE LA JUGADORA Y DEL PC SEAN MENORES A LA VARIABLE PUNTOS TOTALES (3) ....
    contador_rondas += 1 #SUMALE 1 RONDA
    print(f"------ RONDA {contador_rondas} ------") #IMPRIME EL NUMERO DE RONDA QUE COINCIDA

    jugadora_elige = eleccion_jugadora()  # NOMBRE DE VARIABLE REHUTILIZADA DE LA FUNCIÃ“N (VARIABLE LOCAL) 
    pc_elige = random.choice(opciones)    #PC ELIGE SU OPCION ALEATORIAMENTE
    print(f"ğŸ‘©ğŸ»: {jugadora_elige}  ğŸ¤–: {pc_elige}\n") #MUESTRA LO QUE ESCOGIÃ“ CADA JUGADOR

    if jugadora_elige == pc_elige: #SI ELIGEN LA MISMA OPCION EMPATE
        print("EMPATE ğŸ¤ğŸ¼\n")       #MUESTRA 
    elif opcion_correcta[jugadora_elige] == pc_elige:    #LA JUGADORA GANA SI SU CLAVE EN EL DICCIONARIO COINDICE CON EL VALOR ELEGIDO POR EL PC 
        print(f"{nombre_jugadora}: GANA LA RONDA ğŸ‰\n")  #MUESTRA JUGADORA GANA
        victorias_jugadora += 1 #SUMALE UNA VICTORIA
    else:
        print("ğŸ¤– GANA LA RONDA \n") #DE LO CONTRARIO MUESTRA QUE PC GANA LA RONDA 
        victorias_pc += 1             #SUMALE UN PUNTO


#------------- CONTEO DE RONDAS -------------
print("******************************") #DISEÃ‘O

if victorias_jugadora == puntos_totales: # SI LAS VICTORIAS DE LA JUGADORA ES IGUAL A 3 ANUNCIALA COMO GANADORA
    print(F"{nombre_jugadora} HAS GANADO EL JUEGO ğŸ…\n") #MUESTRA QUE JUGADORA HA GANADO
else:
    print("EL PC HA GANADO â˜¹ï¸ INTÃ‰NTALO DE NUEVO \n")    #DE LO CONTRARIO MUESTRA QUE PC HA GANADO


#-------------- RESULTADO FINAL --------------  
print(f"PUNTUACIÃ“N FINAL: ğŸ‘©ğŸ»: {victorias_jugadora} | ğŸ¤–: {victorias_pc}\n") # CONTEO FINAL DE PUNTOS


#----------------- DESPEDIDA -----------------
print("****** GRACIAS POR JUGAR ******")
```
